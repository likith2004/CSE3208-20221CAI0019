```python
 RomaniaMap=dict()

 RomaniaMap['A']=[('S', None, 1), ('T', None, 1), ('Z', None, 1)]
 RomaniaMap['Z']=[('A', None, 1), ('O', None, 1)],
 RomaniaMap['O']=[('S', None, 1), ('Z', None, 1)],
 RomaniaMap['S']=[('A', None, 1), ('F', None, 1), ('O', None, 1), ('R', None, 1)],
 RomaniaMap['T']=[('A', None, 1), ('L', None, 1)],
 RomaniaMap['L']=[('M', None, 1), ('T', None, 1)],
 RomaniaMap['M']=[('D', None, 1), ('L', None, 1)],
 RomaniaMap['D']=[('C', None, 1), ('M', None, 1)],
 RomaniaMap['C']=[('D', None, 1), ('P', None, 1), ('R', None, 1)],
 RomaniaMap['R']=[('C', None, 1), ('P', None, 1), ('S', None, 1)],
 RomaniaMap['F']=[('B', None, 1), ('S', None, 1)],  # Fagaras
 RomaniaMap['P']=[('B', None, 1), ('C', None, 1), ('R', None, 1)],
 RomaniaMap['B']=[('F', None, 1), ('G', None, 1), ('P', None, 1), ('U', None, 1)],
 RomaniaMap['G']=[('B', None, 1)],
 RomaniaMap['U']=[('B', None, 1), ('H', None, 1), ('V', None, 1)],
 RomaniaMap['H']=[('E', None, 1), ('U', None, 1)],
 RomaniaMap['E']=[('H', None, 1)],
 RomaniaMap['I']=[('N', None, 1), ('V', None, 1)],
 RomaniaMap['N']=[('I', None, 1)]




romaniaMap = {
    'Arad': ['Sibiu', 'Zerind', 'Timisoara'],
    'Zerind': ['Arad', 'Oradea'],
    'Oradea': ['Zerind', 'Sibiu'],
    'Sibiu': ['Arad', 'Oradea', 'Fagaras', 'Rimnicu'],
    'Timisoara': ['Arad', 'Lugoj'],
    'Lugoj': ['Timisoara', 'Mehadia'],
    'Mehadia': ['Lugoj', 'Drobeta'],
    'Drobeta': ['Mehadia', 'Craiova'],
    'Craiova': ['Drobeta', 'Rimnicu', 'Pitesti'],
    'Rimnicu': ['Sibiu', 'Craiova', 'Pitesti'],
    'Fagaras': ['Sibiu', 'Bucharest'],
    'Pitesti': ['Rimnicu', 'Craiova', 'Bucharest'],
    'Bucharest': ['Fagaras', 'Pitesti', 'Giurgiu', 'Urziceni'],
    'Giurgiu': ['Bucharest'],
    'Urziceni': ['Bucharest', 'Vaslui', 'Hirsova'],
    'Hirsova': ['Urziceni', 'Eforie'],
    'Eforie': ['Hirsova'],
    'Vaslui': ['Iasi', 'Urziceni'],
    'Iasi': ['Vaslui', 'Neamt'],
    'Neamt': ['Iasi']
}


def take_romania_map_input():
    RomaniaMap = dict()

    print("Enter the connections between cities in the format 'A S 1' (meaning A is connected to S with a cost of 1).")
    print("Enter 'done' when you're finished.")

    while True:
        user_input = input("Enter connection: ")
        if user_input.lower() == 'done':
            break

        city1, city2, cost = user_input.split()
        cost = int(cost)

        if city1 not in RomaniaMap:
            RomaniaMap[city1] = []

        RomaniaMap[city1].append((city2, None, cost))


        if city2 not in RomaniaMap:
            RomaniaMap[city2] = []

        RomaniaMap[city2].append((city1, None, cost))

    return RomaniaMap

def take_romania_full_map_input():
    romaniaMap = dict()

    print("Enter the full city connections in the format 'Arad Sibiu Timisoara Zerind'.")
    print("Each line should include the city followed by its connected cities.")
    print("Enter 'done' when you're finished.")

    while True:
        user_input = input("Enter connection: ")
        if user_input.lower() == 'done':
            break

        parts = user_input.split()
        city = parts[0]
        neighbors = parts[1:]

        romaniaMap[city] = neighbors

    return romaniaMap


romania_map_abbrev = take_romania_map_input()
print("RomaniaMap (abbreviated):", romania_map_abbrev)

romania_map_full = take_romania_full_map_input()
print("romaniaMap (full names):", romania_map_full)




```
